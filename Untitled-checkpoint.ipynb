### I denne prosjektoppgaven 칮nsker jeg 친 se p친 ankomstene av enslige mindre친rige asyls칮kere (EMA) i 2017. Tallene p친 ankomster per m친ned er hentet fra 游댕[UDI](https://www.udi.no/statistikk-og-analyse/statistikk/asylsoknader-enslige-mindrearige-asylsokere-2017/), men kj칮nn og alder er fiktive.

#### Denne grafen viser fordelingen av EMA-ankomster til Norge i 2017, visualisert i et stolpediagram. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_excel("EMA2017.xlsx") #for 친 lese excel-filen jeg har laget med data fra UDI og fiktiv data.

maaned = data['Maaned'].values
u_dag = data['ukedag'].values
land = data['land'].values
kjonn = data['kjonn'].values
alder = data['alder'].values

EMA_mnd = data['Maaned'].value_counts()

month_order = ['januar', 'februar', 'mars', 'april', 'mai', 'juni', 'juli', 'august', 'september', 'oktober', 'november', 'desember']
EMA_mnd = EMA_mnd.reindex(month_order, fill_value=0)

#Under lager jeg stolpediagrammet med farger, titler og ruterett. 
plt.figure(figsize=(10, 6))
EMA_mnd.plot(kind='bar', color='green', edgecolor='black')


plt.title('Antall EMA per m친ned i 2017', fontsize=16)
plt.xlabel('') #Jeg 칮nsker ikke en tittel p친 x-aksen fordi det holder med navnene p친 m친nedene.
plt.ylabel('Antall EMA', fontsize=14)
plt.xticks(rotation=45) #for 친 'tilte' titlene p친 x-aksen.
plt.grid(axis='y', linestyle='--', alpha=0.7) #

plt.tight_layout()
plt.show()

```

#### N친 칮nsker jeg 친 finne ut hvilke ukedager som har de h칮yeste ankomstene av EMA. Dette kan v칝re praktisk 친 vite for 1.linje i utlendingsforvaltningen n친r de skal planlegge n친r behovet for personell, tolker og representanter er st칮rst. 

```python
ukedag_antall = data['ukedag'].value_counts()
#jeg lager en egendefinert funksjon for 친 lage et kakediagram slik at jeg kan bruke det senere i programmet ogs친.
def egendefinert_kake(data, tittel, colors=None):
    plt.figure(figsize=(8, 8))
    data.plot(kind='pie', autopct='%1.f%%', startangle=90, colors=colors, shadow=False) #I denne koden ligger utregningen av prosent i 'autopct' og jeg har valgt '1.' for at det ikke skal v칝re desimaltall fordi det er snakk om mennesker som er 1.
    plt.title(tittel, fontsize=16)
    plt.ylabel('') #for 친 ikke ha en y-tittel
    plt.show()


egendefinert_kake(
    data = ukedag_antall,
    tittel = 'S칮knader fra EMA fordelt p친 ukedager',
    colors=plt.cm.Paired.colors,
)
    


#For 친 sjekke at 'autopct' har regnet riktig prosent skriver jeg inn utregningen ogs친
ukedag_prosent = ((ukedag_antall / ukedag_antall.sum()) * 100).round() #Rundet opp for 친 f친 et heltall. Her kunne jeg ogs친 endret typen til 'integer', men vil da f친 kun heltallet, og ikke avrundet til n칝rmeste heltall. 

#N친r dette skriver ut 칮nsker jeg 친 f친 dagene i riktig rekkef칮lge.
riktig_rekkef칮lge = ['mandag', 'tirsdag', 'onsdag', 'torsdag', 'fredag', 'lordag', 'sondag']

ukedag_prosent = ukedag_prosent.reindex(riktig_rekkef칮lge)

print(ukedag_prosent)


```

### Jeg 칮nsker 친 finne ut hvordan EMA fordeler s

```python
file_path = ('EMA2017.xlsx')
data = pd.read_excel(file_path) #henter og leser datasettet fra excel.

kontinenter = ['Afrika', 'Europa', 'Asia', 'Nord Amerika', 'Sor Amerika', 'Australia'] #Legger til kontinentene.
landegrupper = {
    'Afrika': ['ERI', 'MAR', 'SOM', 'NGA', 'ETH', 'SDN', 'ALG', 'EGY', 'GIN', 'DRC', 'CIV'],
    'Asia': ['AFG', 'SYR', 'IRQ', 'IRN', 'LBN', 'PHL', 'CHN'],
    'Europa': ['ALB', 'BIH'],
    'Nord Amerika': ['USA'],
    'Sor Amerika': [],
    'Australia': [],
} #Fordeler landene i datasettet til det riktige kontinentet.

kontinent_dictionary = {land: kontinent for kontinent, lands in landegrupper.items() for land in lands} #Lager en dictionary for landene og kontinentene de tilh칮rer.

data['land_til_kontinent'] = data['land'].map(kontinent_dictionary) #for 친 endre fra land til kontinent som fodelt i koden over. 

kontinenter = data['land_til_kontinent'].value_counts() #Teller landene i datasettet.

#Jeg vil ha et diagram som viser fordelingen av kontinenter EMAene kommer fra i prosent.
import matplotlib.pyplot as plt
kontinent_prosent = (kontinenter / kontinenter.sum()) * 100 #regne ut prosent
kontinent_prosent.plot(kind='barh', color=plt.cm.Paired.colors, figsize=(10, 6))
plt.title('Fordelingen av hvilke kontinenter EMA kommer fra', fontsize=16)
plt.ylabel('')
plt.xlabel('Prosentandel')

plt.show()


    
               
    
```
